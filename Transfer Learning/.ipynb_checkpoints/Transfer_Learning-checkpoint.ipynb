```python
def display_features_map(vgg16,img,img_name,number_channels_to_display=10):
    vgg16.eval()

    first_conv_layer = vgg16.features[0]

    # Preprocess the image
    preprocess = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    img_prepro = preprocess(img)
    img_prepro = img_prepro.unsqueeze(0)

    # Perform inference
    with torch.no_grad():
        activations_map = first_conv_layer(img_prepro)

    # Visualize the activation maps
    num_channels = activations_map.shape[1]

    print(f"Features Map of {img_name}'s image with {number_channels_to_display} channels : ")


    plt.figure(figsize=(15, 10))
    number_channels_to_display = 10
    for i in range(num_channels):
        # just to avoid the case where the i become 10 and i+1 11 and then error in the subplot
        if i == number_channels_to_display:
          break
        plt.subplot(1, number_channels_to_display, i + 1)
        # batch 0 (unique batch)
        plt.imshow(activations_map[0, i].numpy(), cmap='viridis')
        plt.axis('off')
        plt.title(f'Channel {i + 1}')

    plt.tight_layout()
    plt.show()
```

```python
# display the features maps of the images
for img,img_name in zip(images,images_names):
  display_features_map(vgg16,img,img_name)
```

Interpreting the activation maps allows us the understanding of the patterns and features that specific channels in the first convolutional layer respond to. Each channel in the convolutional layer corresponds to a filter or feature detector, and the activation map represents the response of that filter to different patterns in our input image.

This first convolutional layer captures low-level features such as edges, corners, and simple textures. Each channel is sensitive to a different type of edge or texture.

Bright areas in this activation maps indicate regions in the input image where the corresponding filter detected a feature. Dark areas suggest that the filter did not respond strongly to that region.

```python
class VGG16relu7(nn.Module):
    def __init__(self,vgg16):
      super(VGG16relu7, self).__init__()
      # Copy the entire convolutional part
      self.features = nn.Sequential( *list(vgg16.features.children()))
      # Keep a piece of the classifier: -2 to stop at relu7
      self.classifier = nn.Sequential(*list(vgg16.classifier.children())[:-2])
    def forward(self, x):
      x = self.features(x)
      x = x.view(x.size(0), -1)
      x = self.classifier(x)
      return x
```

```python

PRINT_INTERVAL = 50
CUDA = True

def get_dataset(batch_size, path):

    # Cette fonction permet de recopier 3 fois une image qui
    # ne serait que sur 1 channel (donc image niveau de gris)
    # pour la "transformer" en image RGB. Utilisez la avec
    # transform.Lambda
    def duplicateChannel(img):
        img = img.convert('L')
        np_img = np.array(img, dtype=np.uint8)
        np_img = np.dstack([np_img, np_img, np_img])
        img = Image.fromarray(np_img, 'RGB')
        return img

    train_dataset = datasets.ImageFolder(path+'/train',
        transform=transforms.Compose([ # Pré-traitement à faire
            transforms.Resize((224,224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ]))
    val_dataset = datasets.ImageFolder(path+'/test',
        transform=transforms.Compose([ # Pré-traitement à faire
            transforms.Resize((224,224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ]))
    train_loader = torch.utils.data.DataLoader(train_dataset,
                        batch_size=batch_size, shuffle=False, pin_memory=CUDA, num_workers=2)
    val_loader = torch.utils.data.DataLoader(val_dataset,
                        batch_size=batch_size, shuffle=False, pin_memory=CUDA, num_workers=2)

    return train_loader, val_loader
```

```python
def extract_features(data, model,verbose=False):
    # init features matrices
    N = len(data)
    X = []
    y = []

    for i, (input, target) in enumerate(data):
        if i % PRINT_INTERVAL == 0:
            if verbose : print('Batch {0:03d}/{1:03d}'.format(i, len(data)))
        if CUDA:
            input = input.cuda()
            model = model.cuda()

        # Feature extraction à faire
        with torch.no_grad():
          X.extend([ vector for vector in model(input)])
          y.extend([label for label in target])

        # Empty GPU cache
        torch.cuda.empty_cache()

    X = torch.stack(X,dim=0)
    y = torch.stack(y,dim=0)

    return X, y


def main(path="15SceneData", batch_size=8):
    print('Instanciation de VGG16')
    vgg16 = models.vgg16(pretrained=True)

    print('Instanciation de VGG16relu7')
    # Remplacer par le modèle par un réseau tronqué pour faire de la feature extraction
    # On créera une nouvelle classe VGG16relu7 ici
    model = VGG16relu7(vgg16)
    model.eval()
    if CUDA: # si on fait du GPU, passage en CUDA
        cudnn.benchmark = True
        model = model.cuda()

    # On récupère les données
    print('Récupération des données')
    train, test = get_dataset(batch_size, path)

    # Extraction des features
    print('Feature extraction')
    X_train, y_train = extract_features(train, model)
    X_test, y_test = extract_features(test, model)

    # Apprentissage et évaluation des SVM
    print('Apprentissage des SVM')
    svm = LinearSVC(C=1.0)
    X_train = X_train.cpu()
    X_test = X_test.cpu()
    svm.fit(X_train,y_train)
    accuracy = svm.score(X_test,y_test)
    print('Accuracy = %f' % accuracy)
```

```python
main("15SceneData",8)
```

## Comparative Analysis of Feature Extractors from VGG16

```python
class Truncted_VGG16(nn.Module):
    def __init__(self,vgg16,index_layer_stop=7,shape_layer=(14,14,512)):

      # the index layer stop is the index choosen from the end of the vgg16 (1 --> last layer , 10 middle layer , etc ... )

      super(Truncted_VGG16, self).__init__()
      # Copy the convolutional part
      self.features = nn.Sequential( *list(vgg16.features.children())[:-index_layer_stop])

      # change only the first layer to have to same shape as the output of the features part
      input_shape = shape_layer[0]*shape_layer[1]*shape_layer[2]
      self.classifier = nn.Sequential(
          nn.Linear(in_features=input_shape, out_features=4096, bias=True),
          *list(vgg16.classifier.children())[1:-2]
      )

    def forward(self, x):
      x = self.features(x)
      x = x.view(x.size(0), -1)
      x = self.classifier(x)
      return x
```

```python
def GlobalLearning(model,path="15SceneData", batch_size=8):

  model.eval()
  if CUDA: # si on fait du GPU, passage en CUDA
    cudnn.benchmark = True
    model = model.cuda()

  # On récupère les données
  train, test = get_dataset(batch_size, path)

  # Extraction des features
  X_train, y_train = extract_features(train, model)
  X_test, y_test = extract_features(test, model)

  # Apprentissage et évaluation des SVM
  svm = LinearSVC(C=1.0)
  X_train = X_train.cpu()
  X_test = X_test.cpu()
  svm.fit(X_train,y_train)
  accuracy = svm.score(X_test,y_test)

  return accuracy
```

## Tuning the SVM parameter C

```python
def GlobalLearning_with_Tuning(model,path="15SceneData", batch_size=8):

  model.eval()
  if CUDA: # si on fait du GPU, passage en CUDA
    cudnn.benchmark = True
    model = model.cuda()

  # On récupère les données
  train, test = get_dataset(batch_size, path)

  # Extraction des features
  X_train, y_train = extract_features(train, model)
  X_test, y_test = extract_features(test, model)

  param_grid = {'C': [0.0001,0.001, 0.01, 0.1, 1]}

  svm = LinearSVC()

  grid_search = GridSearchCV(svm, param_grid, cv=5)
  X_train = X_train.cpu()
  X_test = X_test.cpu()
  grid_search.fit(X_train, y_train)

  best_svm = grid_search.best_estimator_
  accuracy = best_svm.score(X_test, y_test)

  return accuracy
```

## Replacing the Last Layer and Training on the 15 Scene Dataset

```python
class CustomVGG16(nn.Module):

  def __init__(self, vgg16,num_classes=15,*args, **kwargs) -> None:
    super().__init__(*args, **kwargs)

    self.features = nn.Sequential(*list(vgg16.features.children()))
    self.avgpool = vgg16.avgpool
    self.classifier = nn.Sequential(*list(vgg16.classifier.children())[:-1],
                      nn.Linear(4096, num_classes))

    #training only on the added layer (the last layer)
    for param in self.classifier[-1].parameters():
      param.requires_grad = True


  def forward(self,x):
    x = self.features(x)
    x = self.avgpool(x)
    x = x.view(x.size(0), -1)
    x = self.classifier(x)

    return x
```

```python
vgg16_15scene = CustomVGG16(vgg16,num_classes=15)
```

```python
train, test = get_dataset(8, "15SceneData")
```

```python
model = vgg16_15scene

optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
criterion = nn.CrossEntropyLoss()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

train_losses  = []
validation_accuracy = []

num_epochs = 20
for epoch in range(num_epochs):
    model.train()

    for inputs, labels in train:
        temp_losses = []
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)

        loss = criterion(outputs, labels)

        loss.backward()
        temp_losses.append(loss.item())
        optimizer.step()
    train_losses.append(np.mean(temp_losses))

    model.eval()
    correct = 0
    total = 0


    with torch.no_grad():
        for inputs, labels in test:
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)

            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = correct / total
    validation_accuracy.append(accuracy)
    print(f'Epoch {epoch + 1}/{num_epochs}, Validation Accuracy: {accuracy:.4f}')
# torch.save(model.state_dict(), 'fine_tuned_vgg16.pth')
```

## Dimensionality reduction

```python
def LearningWithDimReduction(model,model_reduction_dim,X_train,y_train,X_test,y_test,reduction=True):

  model.eval()
  if CUDA: # si on fait du GPU, passage en CUDA
    cudnn.benchmark = True
    model = model.cuda()

  # dimension reduction
  X_train = X_train.cpu()
  X_test = X_test.cpu()

  if reduction:
    tac_dim = time.time()
    X_train_reduc = model_reduction_dim.fit_transform(X_train)
    tic_dim = time.time()
    dim_reduc_time = np.round(tic_dim-tac_dim,4)
    print(f"Dimensionality reduction ends after {dim_reduc_time}s")

    X_test_reduc = model_reduction_dim.transform(X_test)
  else:
    X_train_reduc = X_train
    X_test_reduc = X_test

  # Apprentissage et évaluation des SVM
  svm = LinearSVC(C=0.001)
  tac_svm = time.time()
  svm.fit(X_train_reduc,y_train)
  tic_svm = time.time()
  svm_time = np.round(tic_svm-tac_svm,4)
  print(f"SVM training ends after {svm_time}s")
  accuracy = svm.score(X_test_reduc,y_test)

  if reduction :
    return accuracy , dim_reduc_time , svm_time
  else:
    return accuracy , svm_time
```

```python
vgg16_relu7= VGG16relu7(vgg16)

path="15SceneData"
batch_size=8
# On récupère les données
train, test = get_dataset(batch_size, path)

# Extraction des features
X_train, y_train = extract_features(train, vgg16_relu7)
X_test, y_test = extract_features(test, vgg16_relu7)
X_train = X_train.cpu()
X_test = X_test.cpu()
```

```python
# PCA
pca = PCA(n_components=1000)
pca_acc, pca_reduc_time , pca_svm_time = LearningWithDimReduction(vgg16_relu7,pca,X_train,y_train,X_test,y_test)
print(pca_acc , pca_reduc_time , pca_svm_time)
```

```python
# Kernel rbf PCA
kpca_rbf = KernelPCA(n_components=1000, kernel='rbf', gamma=15)
kpca_rbf_acc, kpca_rbf_reduc_time , kpca_rbf_svm_time = LearningWithDimReduction(vgg16_relu7,kpca_rbf,X_train,y_train,X_test,y_test)
print(kpca_rbf_acc , kpca_rbf_reduc_time , kpca_rbf_svm_time)
```

```python
# Kernel Polynomial PCA
kpca_poly = KernelPCA(n_components=1000, kernel='poly', gamma=15)
kpca_poly_acc, kpca_poly_reduc_time , kpca_poly_svm_time = LearningWithDimReduction(vgg16_relu7,kpca_poly,X_train,y_train,X_test,y_test)
print(kpca_poly_acc , kpca_poly_reduc_time , kpca_poly_svm_time)
```

We notice that when using a RBF kernel the concervation of features is not present , on the other hand when using a Polynomial kernel is more suitable for our data . We can conclude that the features extraction of our images has polynomial relationships.

```python
class Autoencoder(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(Autoencoder, self).__init__()

        # Encoder
        self.encoder = nn.Sequential(
            nn.Linear(input_size, input_size//2),
            nn.ReLU(),
            nn.Linear(input_size//2, hidden_size),
            nn.ReLU()
        )

        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(hidden_size, input_size//2),
            nn.ReLU(),
            nn.Linear(input_size//2,input_size),
            nn.Sigmoid()  # Sigmoid activation for reconstruction between 0 and 1
        )

    def forward(self, x):
        # Forward pass through the encoder and decoder
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
from enum import auto
input_size = X_train.size(1)
hidden_size = 1000

autoencoder = Autoencoder(input_size, hidden_size)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
autoencoder.to(device)

criterion = nn.MSELoss()
optimizer = optim.Adam(autoencoder.parameters(), lr=0.0001)

num_epochs = 10
loss_train = []
loss_val = []


for epoch in tqdm(range(num_epochs)):
    running_loss = 0.0
    autoencoder.train()
    temp_train_loss = []
    temp_val_loss = []


    for i in range(X_train.size(0)):

        inputs = X_train[i]
        inputs = inputs.to(device)

        optimizer.zero_grad()

        outputs = autoencoder(inputs)

        loss = criterion(outputs, inputs)

        loss.backward()
        optimizer.step()

        temp_train_loss.append(loss.item())

    loss_train.append(np.mean(temp_train_loss))


    autoencoder.eval()
    for j in range(X_val.size(0)):
        with torch.no_grad():

          inputs = X_val[j]
          inputs = inputs.to(device)

          outputs = autoencoder(inputs)

          loss = criterion(outputs, inputs)
          temp_val_loss.append(loss.item())

    loss_val.append(np.mean(temp_val_loss))


    print(f'Epoch [{epoch + 1}/{num_epochs}], Train loss: {loss_train[epoch]:.4f} , Val loss: {loss_val[epoch]:.4f}')

print('Training complete')
```

```python
def LearningWithAutoEncoder(model,autoencoder,X_train,y_train,X_test,y_test):

  model.eval()
  if CUDA:
    cudnn.benchmark = True
    model = model.cuda()
    autoencoder = autoencoder.cuda()


  # autoencoder is trained on cuda
  X_train = X_train.cuda()
  X_test = X_test.cuda()

  # dimension reduction
  tac_dim = time.time()
  X_train_reduc = autoencoder(X_train)
  tic_dim = time.time()
  dim_reduc_time = np.round(tic_dim-tac_dim,4)
  print(f"Dimensionality reduction ends after {dim_reduc_time}s")

  X_test_reduc = autoencoder(X_test)

  # Apprentissage et évaluation des SVM
  X_train_reduc = X_train_reduc.cpu().detach().numpy()
  X_test_reduc = X_test_reduc.cpu().detach().numpy()
  svm = LinearSVC(C=0.001)
  tac_svm = time.time()
  svm.fit(X_train_reduc,y_train)
  tic_svm = time.time()
  svm_time = np.round(tic_svm-tac_svm,4)
  print(f"SVM training ends after {svm_time}s")
  accuracy = svm.score(X_test_reduc,y_test)

  return accuracy , dim_reduc_time , svm_time
```

```python
ae_acc, ae_reduc_time , ae_svm_time = LearningWithAutoEncoder(vgg16_relu7,autoencoder,X_train,y_train,X_test,y_test)
print(ae_acc , ae_reduc_time , ae_svm_time)
```
